```python
# Binary Tree searching.

def search(nums, target):
    from math import floor
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = floor((left + right) / 2)
        if nums[mid] < target:
            left = mid + 1
        elif nums[mid] > target:
            right = mid - 1
        else:
            return mid
    
    return -1 
```

```python
# Tree for checking bad versions.

def isBadVersion(n):
    if n == 2:
        return True
    else:
        return False

def firstBadVersion(n):
    """
    :type n: int
    :rtype: int
    """
    from math import floor
    
    left = 1
    right = n
    
    while left <= right:
        mid = floor((left + right)/2)
        is_bad_mid = isBadVersion(mid)
        is_bad_mid_minus_1 = isBadVersion(mid - 1)
        
        if is_bad_mid and is_bad_mid_minus_1:
            right = mid - 1
        elif not is_bad_mid and not is_bad_mid_minus_1:
            left = mid + 1
        else:
            return mid

#print(firstBadVersion(2))
```

```python

# Binary tree search + insertion index

def searchInsert(nums, target) -> int:
        # Case for returning index
        left = 0
        right = len(nums) - 1
        if target <= min(nums):
            return 0
        elif target > max(nums):
            return len(nums)

        while left <= right:
            mid = floor((left + right) / 2)
            if nums[mid] == target:
                return mid
            elif nums[mid] > target and nums[mid - 1] < target:
                return mid
            elif nums[mid] < target and nums[mid + 1] > target:
                return mid + 1
            elif nums[mid] < target:
                left = mid + 1
            elif nums[mid] > target:
                right = mid - 1
            else:
                return mid

# nums = [1,3]
# target = 3
# print(searchInsert(nums, target))

```
